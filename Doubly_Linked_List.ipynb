```python
class Node:
    def __init__(self,data=None,next=None,previous=None):
        self.data = data
        self.next = next
        self.previous = previous

class DLinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_head(self,data):
        node = Node(data,self.head)
        if self.head:
            self.head.previous = node
            self.head = node
        else:
            self.head = node

    def insert_at_last(self,data):
        node = Node(data)
        if self.head:
            itr = self.head
            while itr.next:
                itr = itr.next
            node.previous = itr
            itr.next = node
        else:
            self.head = node

    def get_head(self):
        return self.head

    def get_tail(self):
        itr = self.head
        lst = ''
        while itr.next:
            itr = itr.next
        return itr

    def print_reverse(self):
        itr = self.get_tail()
        lst = ''
        while itr:
            lst += str(itr.data) + ' -->'
            itr = itr.previous
        return lst

    def __repr__(self):
        itr = self.head
        lst = ''
        while itr:
            lst += str(itr.data) + ' -->'
            itr = itr.next
        return lst
```

```python
dl = DLinkedList()
dl.insert_at_last(7)
dl.insert_at_last(8)
dl.insert_at_last(9)
dl.insert_at_last(10)
dl.get_tail().previous.data
```
