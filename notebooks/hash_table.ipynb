```python
class HashTable:
    def __init__(self,MAX):
        self.MAX = MAX
        self.arr = [[] for i in range(self.MAX)]

    def get_hash(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX

    def add(self,key,val):
        indx = get_hash(key)
        ## For chaining to avoid collision
        isfound = False
        data = self.arr[indx]
        for loc,pair in enumerate(data):
            if pair[0] == key:
                data[loc] = (key,val)
                isfound = True
                break
        if not isfound:
            self.arr[indx].append((key,val))

    def get(self,key):
        val = ''
        data = self.arr[get_hash(key)]
        for pair in data:
            if pair[0] == key:
                return pair[1]
                break
```
