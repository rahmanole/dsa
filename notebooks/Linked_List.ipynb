```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class LinkedList:
    def __init__(self) -> None:
        self.head = None

    def insert_at_begining(self,data):
        node = Node(data,self.head)
        self.head = node
        

    def insert_at_last(self,data):
        node = Node(data)

        if self.head is None:
            self.head = node
            return
        
        itr = self.head

        while itr.next:
            itr = itr.next

        itr.next = node

    
    def insert_before(self,data,position):
        if self.head is None:
            return 'empty linkedlist'

        node = Node(data)
        itr = self.head
        while itr.next:
            if itr.next.data == position:
                node.next = itr.next
                itr.next = node
                return
            itr = itr.next
            
        return 'position not found'

    def insert_after(self,data,position):
        if self.head is None:
            return 'empty linkedlist'

        node = Node(data)
        itr = self.head
        while itr.data:
            if itr.data == position:
                node.next = itr.next
                itr.next = node
                return
            itr = itr.next

    def remove(self,data):
        itr = self.head
        while itr.data:
            if itr.next.data == data:
                itr.next = itr.next.next
                return
            itr = itr.next


    def get_head(self):
        return self.head
    
    def get_tail(self):
        itr = self.head
        while itr.next:
            itr = itr.next
        return itr.data

    def length(self):
        counter = 0
        itr = self.head
        while itr:
            counter += 1
            itr = itr.next
        return counter

    def print(self):
        itr = self.head
        while itr:
            print(itr.data)
            itr = itr.next


    def __repr__(self):
        if self.head is None:
            return 'list is empty'
        else: 
            itr = self.head
            data = ''
            while itr:
                data += str(itr.data) + '-->'
                itr = itr.next
            return data
```

```python
l = LinkedList()
l.insert_at_last(1)
l.insert_at_last(2)
l.insert_at_last(3)
l.insert_at_last(4)
l
```

```python
def reverseList(head):
    prev = None
    original_head = head
    while head:
        curr = head
        head = head.next
        curr.next = prev
        prev = curr
        
    while original_head is not None:
        if original_head.data != prev.data:
            return False
        original_head = original_head.next
        prev = prev.next
    return True
```
