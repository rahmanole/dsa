### Queue implementation using list

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def put(self,x):
        self.queue.append(x)
    
    def pop(self):
         return self.queue.pop(0) if len(self.queue)>0 else -1
    
    def empty(self):
        return False if len(self.queue)>0 else True
    
    
```

```python
q = Queue()
q.put(1)
q.put(2)
q.put(3)
q.put(4)
q.pop()
```

### Queue Module of Python
*   When information must be exchanged safely between multiple threads.
*   The module implements three types of queue:
    *   FIFO queue
    *   LIFO queue
    *   Priority queue, the entries are kept sorted and the lowest valued entry
