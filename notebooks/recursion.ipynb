```python
## Find sum till nth number
def sum(n):
    if n==1:
        return 1
    else:
        return n + sum(n-1)
```

```python
sum(10)
```

```python
## Fibonacci number
def p(n):
    if n==1 or n==2:
        return 1
    elif n<=0:
        return 'number should be greater than or equal to 0'
    else:
        return p(n-2) + p(n-1)
```

```python
p(8)
```

```python
## Factorial 
def fact(n):
    if n<0:
        return 'Enter valid number'
    elif n==0 or n==1:
        return 1
    else:
        return fact(n-1)*n
```

```python
fact(6)
```

```python
## Find sum of digits
def digit_sum(x):
    if x < 10:
        return x
    else:
        mod = x%10
        x = x//10
        v = mod + digit_sum(x) 
        return digit_sum(v)
```

```python
digit_sum(2349)
```

```python
def addDigits(num: int) -> int:
    if num < 10:
        return num
    else:
        mod = num%10
        num = num//10
        digit_sum = mod + addDigits(num)
        return addDigits(digit_sum)
```

```python
addDigits(6787)
```

```python
def gcd(a,b):
    if b == 0:
        return a
    else:
        return gcd(b,a%b)
```

```python
print(gcd(57,18))
```
